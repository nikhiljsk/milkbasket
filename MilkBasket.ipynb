```python
import pandas as pd
import time
st = time.time()
df = pd.read_csv("./temp_bq_data/hackathon_data.csv", header='infer')
print("took", time.time()-st)
```

```python
df['added_date'] = df.apply(lambda row: row['product_addedtobasket_on'].split(' ')[0], axis=1)
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['added_date'] = pd.to_datetime(df['added_date'])
df['Difference'] = (df['order_date'] - df['added_date']).dt.days
```

```python
df['id'] = df['product_id']
df = df.groupby(['product_id']).mean()
```

```python
import numpy as np
df['Difference'] = df.Difference.apply(np.round)
```

```python
df_new = df[df['Difference']>=5]
df_new = df_new.sort_values(by=['Difference'], ascending=False, inplace=False)
```

```python
df_new.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt = plt.figure(figsize=(20,20))
sns.barplot(x=df_new['id'], y=df_new['Difference'])
plt.show()
```
